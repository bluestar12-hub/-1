````python
# @title 3. –ò–º–ø–æ—Ä—Ç –±–∏–±–ª–∏–æ—Ç–µ–∫
import torch
import torch.nn as nn
import torch.optim as optim
from torch.utils.data import Dataset, DataLoader
import json
from transformers import AutoTokenizer, AutoModelForCausalLM
from tqdm import tqdm
import re
import os
import numpy as np
import warnings
warnings.filterwarnings('ignore')

print("‚úÖ –ë–∏–±–ª–∏–æ—Ç–µ–∫–∏ –∏–º–ø–æ—Ä—Ç–∏—Ä–æ–≤–∞–Ω—ã")
````

````python
# @title 4. –ö–æ–Ω—Ñ–∏–≥—É—Ä–∞—Ü–∏—è –º–æ–¥–µ–ª–∏
class Config:–æ
    model_name = "distilgpt2"
    max_length = 512
    batch_size = 4
    epochs = 5  # –£–º–µ–Ω—å—à–∏–ª –¥–ª—è Colab
    learning_rate = 5e-5
    device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
    max_new_tokens = 256
    temperature = 0.7
    top_p = 0.9
    dataset_path = "mbpp.jsonl"

config = Config()

print("=" * 60)
print("–ö–û–ù–§–ò–ì–£–†–ê–¶–ò–Ø")
print("=" * 60)
print(f"üì± –£—Å—Ç—Ä–æ–π—Å—Ç–≤–æ: {config.device}")
print(f"ü§ñ –ú–æ–¥–µ–ª—å: {config.model_name}")
print(f"üìä Batch size: {config.batch_size}")
print(f"üîÑ –≠–ø–æ—Ö–∏: {config.epochs}")
````

````python
# @title 5. –ö–ª–∞—Å—Å –¥–∞—Ç–∞—Å–µ—Ç–∞
class MBPPDataset(Dataset):
    def __init__(self, file_path, max_length=512, split='train', val_split=0.1):
        self.max_length = max_length
        self.split = split

        # –ó–∞–≥—Ä—É–∑–∫–∞ –¥–∞—Ç–∞—Å–µ—Ç–∞
        self.examples = self._load_dataset(file_path)

        # –ò–Ω–∏—Ü–∏–∞–ª–∏–∑–∞—Ü–∏—è —Ç–æ–∫–µ–Ω–∏–∑–∞—Ç–æ—Ä–∞
        self.tokenizer = AutoTokenizer.from_pretrained(config.model_name)
        if self.tokenizer.pad_token is None:
            self.tokenizer.pad_token = self.tokenizer.eos_token

        # –ü–æ–¥–≥–æ—Ç–æ–≤–∫–∞ –ø–∞—Ä –∫–æ–¥-—Ç–µ—Å—Ç
        self.pairs = self._prepare_pairs()

        # –†–∞–∑–¥–µ–ª–µ–Ω–∏–µ –Ω–∞ train/val
        self._split_data(val_split)
        print(f"‚úÖ {split}: {len(self.indices)} –ø—Ä–∏–º–µ—Ä–æ–≤")

    def _load_dataset(self, file_path):
        examples = []
        try:
            with open(file_path, 'r', encoding='utf-8') as f:
                for line in f:
                    try:
                        examples.append(json.loads(line.strip()))
                    except:
                        continue
        except:
            pass
        return examples

    def _prepare_pairs(self):
        pairs = []
        for ex in self.examples:
            try:
                desc = ex.get('text', '')
                code = ex.get('code', '')
                tests = ex.get('test_list', [])

                if not code or not tests:
                    continue

                code = code.strip()

                prompt = f"# –ó–∞–¥–∞—á–∞: {desc}\n\n"
                prompt += f"# –§—É–Ω–∫—Ü–∏—è:\n```python\n{code}\n```\n\n"
                prompt += f"# –ù–∞–ø–∏—à–∏—Ç–µ unit-—Ç–µ—Å—Ç—ã –¥–ª—è —ç—Ç–æ–π —Ñ—É–Ω–∫—Ü–∏–∏:\n```python\n"

                formatted_tests = self._format_tests(tests)

                pairs.append({
                    'prompt': prompt,
                    'test': formatted_tests + "\n```",
                    'desc': desc,
                    'code': code,
                    'tests': tests
                })

            except:
                continue
        return pairs

    def _format_tests(self, tests):
        return "\n".join([f"# Test {i+1}\n{test.strip()}" for i, test in enumerate(tests)])

    def _split_data(self, val_split):
        indices = list(range(len(self.pairs)))
        np.random.seed(42)
        np.random.shuffle(indices)
        split = int(len(self.pairs) * (1 - val_split))
        self.indices = indices[:split] if self.split == 'train' else indices[split:]

    def __len__(self):
        return len(self.indices)

    def __getitem__(self, idx):
        pair = self.pairs[self.indices[idx]]
        full_text = pair['prompt'] + pair['test'] + self.tokenizer.eos_token

        encoding = self.tokenizer(
            full_text,
            max_length=self.max_length,
            padding='max_length',
            truncation=True,
            return_tensors='pt'
        )

        prompt_len = len(self.tokenizer.encode(pair['prompt']))

        input_ids = encoding['input_ids'].squeeze(0)
        attention_mask = encoding['attention_mask'].squeeze(0)

        labels = input_ids.clone()
        labels[:prompt_len] = -100

        return {
            'input_ids': input_ids,
            'attention_mask': attention_mask,
            'labels': labels,
            'prompt': pair['prompt']
        }
````

````python
# @title 6. –ó–∞–≥—Ä—É–∑–∫–∞ –¥–∞–Ω–Ω—ã—Ö
print("=" * 60)
print("–ó–ê–ì–†–£–ó–ö–ê –î–ê–ù–ù–´–•")
print("=" * 60)

train_dataset = MBPPDataset(config.dataset_path, split='train')
val_dataset = MBPPDataset(config.dataset_path, split='val')

train_loader = DataLoader(train_dataset, batch_size=config.batch_size, shuffle=True)
val_loader = DataLoader(val_dataset, batch_size=config.batch_size, shuffle=False)

print(f"\nüìä Train: {len(train_dataset)}")
print(f"üìä Val: {len(val_dataset)}")
````

````python
# @title 7. –ó–∞–≥—Ä—É–∑–∫–∞ –º–æ–¥–µ–ª–∏
print("=" * 60)
print("–ó–ê–ì–†–£–ó–ö–ê –ú–û–î–ï–õ–ò")
print("=" * 60)

model = AutoModelForCausalLM.from_pretrained(config.model_name)
model.to(config.device)
optimizer = optim.AdamW(model.parameters(), lr=config.learning_rate)

print(f"‚úÖ –ú–æ–¥–µ–ª—å –∑–∞–≥—Ä—É–∂–µ–Ω–∞")
print(f"üìä –ü–∞—Ä–∞–º–µ—Ç—Ä–æ–≤: {sum(p.numel() for p in model.parameters()):,}")
````

````python
# @title 8. –§—É–Ω–∫—Ü–∏—è –æ–±—É—á–µ–Ω–∏—è
def train_epoch(model, loader, optimizer, device, epoch):
    model.train()
    total_loss = 0

    pbar = tqdm(loader, desc=f'Epoch {epoch}')
    for batch in pbar:
        input_ids = batch['input_ids'].to(device)
        attention_mask = batch['attention_mask'].to(device)
        labels = batch['labels'].to(device)

        outputs = model(
            input_ids=input_ids,
            attention_mask=attention_mask,
            labels=labels
        )

        loss = outputs.loss

        optimizer.zero_grad()
        loss.backward()
        optimizer.step()

        total_loss += loss.item()
        pbar.set_postfix({'loss': loss.item()})

    return total_loss / len(loader)
````

````python
# @title 9. –û–ë–£–ß–ï–ù–ò–ï
print("=" * 60)
print("–ù–ê–ß–ê–õ–û –û–ë–£–ß–ï–ù–ò–Ø")
print("=" * 60)

for epoch in range(config.epochs):
    print(f"\nüîÑ –≠–ø–æ—Ö–∞ {epoch + 1}/{config.epochs}")

    train_loss = train_epoch(model, train_loader, optimizer, config.device, epoch + 1)
    print(f"üìâ Train Loss: {train_loss:.4f}")

    # Validation
    model.eval()
    val_loss = 0
    with torch.no_grad():
        for batch in tqdm(val_loader, desc='Validation'):
            input_ids = batch['input_ids'].to(config.device)
            attention_mask = batch['attention_mask'].to(config.device)
            labels = batch['labels'].to(config.device)

            outputs = model(
                input_ids=input_ids,
                attention_mask=attention_mask,
                labels=labels
            )

            val_loss += outputs.loss.item()

    val_loss /= len(val_loader)
    print(f"üìä Val Loss: {val_loss:.4f}")
````

````python
# @title 11. –ö–ª–∞—Å—Å –¥–ª—è –≥–µ–Ω–µ—Ä–∞—Ü–∏–∏ —Ç–µ—Å—Ç–æ–≤
class TestGenerator:
    def __init__(self, model):
        self.device = config.device
        self.tokenizer = AutoTokenizer.from_pretrained(config.model_name)
        if self.tokenizer.pad_token is None:
            self.tokenizer.pad_token = self.tokenizer.eos_token
        self.model = model
        self.model.eval()

    def generate(self, code, desc=None):
        if desc:
            prompt = f"# –ó–∞–¥–∞—á–∞: {desc}\n\n"
            prompt += f"# –§—É–Ω–∫—Ü–∏—è:\n```python\n{code}\n```\n\n"
            prompt += f"# –¢–µ—Å—Ç—ã:\n```python\n"
        else:
            prompt = f"# –ö–æ–¥:\n```python\n{code}\n```\n\n# –¢–µ—Å—Ç—ã:\n```python\n"

        inputs = self.tokenizer(prompt, return_tensors='pt', truncation=True, max_length=512).to(self.device)

        with torch.no_grad():
            outputs = self.model.generate(
                inputs.input_ids,
                max_new_tokens=config.max_new_tokens,
                temperature=config.temperature,
                top_p=config.top_p,
                do_sample=True,
                pad_token_id=self.tokenizer.pad_token_id,
                eos_token_id=self.tokenizer.eos_token_id
            )

        result = self.tokenizer.decode(outputs[0][inputs.input_ids.shape[1]:], skip_special_tokens=True)
        if "```" in result:
            result = result.split("```")[0]
        return result.strip()

    def evaluate(self, generated, expected):
        gen = re.findall(r'assert.*', generated)
        exp = re.findall(r'assert.*', '\n'.join(expected))

        if not gen:
            return 0.0

        matches = sum(1 for g in gen if any(e in g or g in e for e in exp))
        return matches / max(len(exp), 1)
````

````python
# @title 12. –¢–ï–°–¢–ò–†–û–í–ê–ù–ò–ï
print("=" * 60)
print("–¢–ï–°–¢–ò–†–û–í–ê–ù–ò–ï –ì–ï–ù–ï–†–ê–¶–ò–ò")
print("=" * 60)

generator = TestGenerator(model)
results = []

for i in range(min(3, len(val_dataset))):
    ex = val_dataset.pairs[val_dataset.indices[i]]

    print(f"\n--- –ü—Ä–∏–º–µ—Ä {i+1} ---")
    print(f"üìù {ex['desc'][:100]}...")
    print(f"üíª –ö–æ–¥:\n{ex['code']}")
    print(f"\n‚úÖ –û–∂–∏–¥–∞–µ–º—ã–µ —Ç–µ—Å—Ç—ã:")
    for t in ex['tests'][:2]:
        print(f"  {t}")

    generated = generator.generate(ex['code'], ex['desc'])
    print(f"\nü§ñ –°–≥–µ–Ω–µ—Ä–∏—Ä–æ–≤–∞–Ω–æ:\n{generated}")

    score = generator.evaluate(generated, ex['tests'])
    print(f"\nüìä –û—Ü–µ–Ω–∫–∞: {score:.2f}")
    results.append(score)
    print("-" * 40)
````
